
# Plotting with categorical data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python

sns.catplot(x="day", y="total_bill", data=tips,kind='strip')
# + kind='strip'
#sns.stripplot('day', 'total_bill', data=tips)
```

```python
sns.catplot('day', 'total_bill', kind='swarm', data=tips)
```

 데이타 순서 정할수 있음 order

```python
sns.catplot(x="smoker", y="tip",order=["No", "Yes"],  data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

```python
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
tips["weekend"] = tips["day"].isin(["Sat", "Sun"])
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", hue="weekend",
            kind="box", dodge=False, data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", bw=.15, cut=4,
            data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", bw=.15, cut=40,
            data=tips)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
# survived: 1, dead: 0
```

```python
titanic.head()
```

```python
titanic[(titanic["class"] == "First") & (titanic["sex"] == "female")& (titanic["survived"] == 1)].count()
```

```python
sns.catplot(y="sex", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic)
```

포인트 선종류 색깔등 지정 가능 

```python
sns.catplot(x="class", y="survived", hue="sex",
            palette={"male": "g", "female": "m"},
            markers=["^", "o"], linestyles=["-", "--"],
            kind="point", data=titanic)
```

사이즈 조절 가능

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c")
```

```python
f, ax = plt.subplots(figsize=(3, 3))
sns.countplot(y="deck", data=titanic, color="c")
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips)
#aspect=.6,그래프 좌우 넓이 
```
